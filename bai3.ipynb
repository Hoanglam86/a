```python
#Bai 3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB

def phan_loai_du_lieu():
    # To d liu mu
    X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
    plt.scatter(X[:, 0], X[:,1], c=y, s=50, cmap='RdBu')
    # p dng m hnh GaussianNB  hc d liu
    model = GaussianNB()
    model.fit(X, y)

    # To 2000 im d liu ngu nhin v d on lp tng ng
    rng = np.random.RandomState(0)
    X_new = rng.rand(2000, 2) * 10 - 5
    y_new = model.predict(X_new)

    # V biu  vi mu sc tng ng vi lp d on
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_new, s=20, cmap='RdBu', alpha=0.1)
    plt.show()

    # In ra trung bnh v  lch chun ca mi nhm sau khi hc d liu
    print("Trung bnh:", model.theta_)
    print(" lch chun:", model.sigma_)


# Gi hm phn loi d liu
phan_loai_du_lieu()
```

```python
import numpy as np


P = np.array([[0.6, 0.3, 0.1, 0.0],
              [0.2, 0.4, 0.3, 0.1],
              [0.1, 0.2, 0.5, 0.2],
              [0.0, 0.3, 0.2, 0.5]])

# Xc nh trng thi ban u
start = np.array([0, 0, 1, 0])  # rainy

# Tnh trng thi sau ba ngy
state = np.linalg.matrix_power(P, 3) @ start


weather = ['Sunny', 'Cloudy', 'Rainy', 'Snowing']
for i, w in enumerate(weather):
    print(f"Xc sut c thi tit {w} sau ba ngy bt u t trng thi ma l {state[i]:.2f}")
```
