```python
#Bai 3
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.datasets import make_blobs
from sklearn.naive_bayes import GaussianNB

def phan_loai_du_lieu():
    # Tạo dữ liệu mẫu
    X, y = make_blobs(100, 2, centers=2, random_state=2, cluster_std=1.5)
    plt.scatter(X[:, 0], X[:,1], c=y, s=50, cmap='RdBu')
    # Áp dụng mô hình GaussianNB để học dữ liệu
    model = GaussianNB()
    model.fit(X, y)

    # Tạo 2000 điểm dữ liệu ngẫu nhiên và dự đoán lớp tương ứng
    rng = np.random.RandomState(0)
    X_new = rng.rand(2000, 2) * 10 - 5
    y_new = model.predict(X_new)

    # Vẽ biểu đồ với màu sắc tương ứng với lớp dự đoán
    plt.scatter(X_new[:, 0], X_new[:, 1], c=y_new, s=20, cmap='RdBu', alpha=0.1)
    plt.show()

    # In ra trung bình và độ lệch chuẩn của mỗi nhóm sau khi học dữ liệu
    print("Trung bình:", model.theta_)
    print("Độ lệch chuẩn:", model.sigma_)


# Gọi hàm phân loại dữ liệu
phan_loai_du_lieu()
```

```python
import numpy as np


P = np.array([[0.6, 0.3, 0.1, 0.0],
              [0.2, 0.4, 0.3, 0.1],
              [0.1, 0.2, 0.5, 0.2],
              [0.0, 0.3, 0.2, 0.5]])

# Xác định trạng thái ban đầu
start = np.array([0, 0, 1, 0])  # rainy

# Tính trạng thái sau ba ngày
state = np.linalg.matrix_power(P, 3) @ start


weather = ['Sunny', 'Cloudy', 'Rainy', 'Snowing']
for i, w in enumerate(weather):
    print(f"Xác suất có thời tiết {w} sau ba ngày bắt đầu từ trạng thái mưa là {state[i]:.2f}")
```
